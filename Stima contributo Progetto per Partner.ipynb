# Carico i file nel sistema

```python
# soggetti per progetto
progettiXpartecipanti = pd.read_excel('Relazione progetti - organizzazioni partecipanti.xlsx')
```

```python
#informazioni sui singoli progetti
progetti = pd.read_excel('Progetti con numero paesi partecipanti.xlsx')
```

```python
progetti
```

```python
'''
print progettiXpartecipanti.columns
print progetti.columns
'''
```

```python
#Voglio un file che raccolga per ogni partner e progetto il contributo totale 
```

```python
risultato = pd.merge(progettiXpartecipanti, progetti[[u'rcn',u'ecMaxContribution']] , left_on=u'projectRcn', right_on=u'rcn')
```

```python
risultato
```

```python
#voglio che il totale del contributo sia diviso per i partner di progetto e quindi mi calcolo il numero di parnter per progetto
```

```python
partner_per_progetto = progettiXpartecipanti.groupby("projectRcn").count().reset_index()
```

```python
partner_per_progetto
```

```python
#voglio aggiungere questa informazione alla mia tabella
```

```python
risultato = pd.merge(risultato , partner_per_progetto , on=u'projectRcn')
```

```python
risultato
```

#clacolo il nuovo campo contributo diviso partecipanti

```python
risultato["stima_quota_per_partner"]= risultato["ecMaxContribution"] /  risultato["org_id_y"]
```

```python
risultato
```

```python
#creo un file csv con all'interno Progetto, partner, quota
```

```python
risultato = risultato[["projectRcn", "ecMaxContribution", "stima_quota_per_partner"]]
```

```python
risultato.to_csv("contributoProgettoXPartner.csv")
```
